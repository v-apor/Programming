```python
# It never raises a keyError. Any key which doesn't exists returns default
```

```python
from collections import defaultdict
```

```python
d = {'k':1}
```

```python
d['k']
```

```python
d['k2']
```

```python
d = defaultdict(object)
```

```python
# returns whatever the object we declared (above) when a key isn't present
d['one']
```

```python
for item in d:
    print(item)
```

```python
d['two'] = 4
```

```python
for item in d:
    print(item)
```

```python
d = defaultdict(lambda : 0)
```

```python
d['one']
```

```python
d['two'] = 2
```

```python
d['two']
```
