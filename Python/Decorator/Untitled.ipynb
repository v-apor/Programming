```python
# Now if there was a way to return a function from a function
```

```python
def hello(name="Apoorv"):
    print("Hello function has been executed")
    
    def greet():
        return '\t This is the greet() function inside hello()!'
    
    def welcome():
        return "\t This is welcome() inside hello()"
    
    print("I am going to return a function!!")
    
    if name == "Apoorv":
        return greet
    else:
        return welcome
```

```python
my_new_func = hello("Apoorv")
```

```python
print(my_new_func())
```

```python
# DECORATOR
```

```python
def new_decorator(original_func):
    
    def wrap_func():
        
        print("Some extra code before the original function")
        
        original_func()
        
        print("Some extra code after the original function")
        
    return wrap_func
```

```python
def func_needs_decorator():
    print("I want to be decorated")
```

```python
func_needs_decorator()
```

```python
decorated_function = new_decorator(func_needs_decorator)   # complex process
```

```python
decorated_function()
```
